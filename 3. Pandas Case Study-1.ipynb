6. In customer table, create a new variable called “customer value segment” using customer value as follows. -
 High Value Segment - > 25000
 Medium Value Segment – Between 10000 and 25000
 Low Value Segment – less than or equal to 10000

```python
import numpy as np

data['customer_value_segment'] = np.where((data['Customer_Value'] > 25000),'High Value Segment',
                                          np.where(((data['Customer_Value'] > 10000) & (data['Customer_Value'] <25000)) ,'Medium Value Segment', 
                                                   np.where((data['Customer_Value'] < 10000),'Low Value Segment',np.nan)))

                                    
```
